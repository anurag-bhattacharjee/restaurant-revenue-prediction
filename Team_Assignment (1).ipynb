```python
Y_Train = Train_Data.iloc[:, -1]
X_Train = Train_Data.iloc[:, :-1]
X_Test = Test_Data.copy()

```

```python
#Do one-hot encoding of categorical variables

combined_Data = pd.concat([X_Train, X_Test], keys=[0,1])

categoricalFeatures = ["City",
"City Group",
"Type"]


#Combine Train and test for one Hot Encoding
#combined_Data = pd.concat([Train_Data,Test_Data], keys=[0,1])

#Do one Hot encoding for categorical features
#combined_Data["y"]=combined_Data["y"].map({"yes":1,"no":0})
combined_Data=pd.get_dummies(combined_Data,columns=categoricalFeatures)
#Updated_Test_Data = pd.get_dummies(Test_Data,columns=categoricalFeatures)
#print(combined_Data['Revenue'])

#Separate Train data and test data
#X_train = combined_Data.xs(0)
#X_test = combined_Data.xs(1)
#X_train.to_csv("/gdrive/My Drive/CS508/updated_train.csv", index = None)
#X_test.to_csv("/gdrive/My Drive/CS508/updated_test.csv", index = None)
#X_train.head()
#X_test.head()
```

```python
Updated_Train_Data.to_csv("/gdrive/My Drive/CS508/Updated_Train_Data.csv", index = None)
Updated_Test_Data.to_csv("/gdrive/My Drive/CS508/Updated_Test_Data.csv", index = None)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Updated_Train_Data, Y_Train, test_size = .25)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Decision Tree Regressor ========================================================================
#CONSTRUCT DEFAULT DECISION TREE AND OBTAIN RESPECTIVE ACCURACY 
dtr = DecisionTreeRegressor()
dtr.fit(X_train, Y_train)
dtr_predict_Train=dtr.predict(X_train)
mean_squared_error(Y_train,dtr_predict_Train)
print("RMSE (training) for Decision Tree:{0:10f}".format(mean_squared_error(Y_train,dtr_predict_Train)))
dtr_predict_Test=dtr.predict(X_test)
print("RMSE (test) for Decision Tree:{0:10f}".format(mean_squared_error(Y_test,dtr_predict_Test)))
```

```python
#CONSTRUCTING DECISION TREE - HYPERPARAMETER TUNING

import time
start_time = time.time()

dtr = DecisionTreeRegressor()
#X_train, X_test, Y_train, Y_test = train_test_split(Updated_Train_Data, Y_Train, test_size = .25,random_state=42)
parameters={'criterion':['squared_error', 'mse', 'absolute_error'],'min_weight_fraction_leaf':[0.01,0.05,0.1], 
            'splitter':['best', 'random'],'min_samples_leaf':[1,100,10]}
dtr_random = RandomizedSearchCV(dtr,parameters,n_iter=25,cv=5)
dtr_random.fit(X_train, Y_train)
grid_parm1=dtr_random.best_params_ 
print(grid_parm1)
dtrRand=DecisionTreeRegressor(**grid_parm1)
dtrRand.fit(X_train,Y_train)
dtrRand_predict=dtrRand.predict(X_test)
print("RMSE for Decision Tree:{0:10f}".format(mean_squared_error(Y_test,dtrRand_predict)))

#get cross-validation report
dtr_cv_score = cross_val_score(dtrRand, X_test, Y_test, cv=4)
print("=== All AUC Scores ===")
print(dtr_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Decision Tree: ",dtr_cv_score.mean())

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
#STACKING MODELS =====================================================================
print("___________________________________________________________________________________________\nEnsemble Methods Predictions using GradientBoosting, RandomForest and Decision Tree Classifier\n")

models = [ DecisionTreeRegressor(), GradientBoostingRegressor(), RandomForestRegressor(), MLPRegressor() ]
      
S_Train, S_Test = stacking(models,                   
                           Updated_Train_Data, Y_Train, Updated_Test_Data,   
                           regression=True, 
     
                           mode='oof_pred_bag', 
       
                           needs_proba=False,
         
                           save_dir=None, 
                                        
                           n_folds=4,
                           
                           shuffle=True,
                                                    
                           verbose=2)

```

```python
#STACKING - CONTRUCT A GRADIENT BOOSTING MODEL==============================
import time
start_time = time.time()

gbr = GradientBoostingRegressor()

Gradient_parameters={'n_estimators':[5,10,20],'learning_rate':[0.01,.1]}

gbr_random = RandomizedSearchCV(gbc,Gradient_parameters,n_iter=10,cv=5)

gbr_random.fit(S_Train, Y_Train)

gbr_grid_parm=gbc_random.best_params_
print(gbr_grid_parm)
gbrRand=GradientBoostingRegressor(**gbr_grid_parm)
gbrRand.fit(S_Train, Y_Train)
gbrRand_predict=gbrRand.predict(S_Train)
#print('Final prediction score for ensemble methods: [%.8f]' % accuracy_score(gbcRand_predict, y_res1))

print("--- %s seconds ---" % (time.time() - start_time))
```
